```python
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
import pandas
```

```python
'''url du site qui m'interesse'''
url='https://www.asos.fr/api/product/search/v2/categories/27108?channel=mobile-web&country=FR&currency=EUR&keyStoreDataversion=p1swt7e-15&lang=fr&limit=100&offset=100&rowlength=2&store=FR'
```

```python
'''recherches de données : les 100 denières nouveautés arrivées '''

response = requests.get(url, headers={"user-agent":'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36'})
results = response.json()
results

flattened_data = json_normalize(results)
flattened_data
```

```python
flattened_data1 = json_normalize(flattened_data.products[0])
flattened_data1

df = pd.DataFrame(flattened_data1)
df
```

```python
'''je supprime les colonnes dont je n'ai pas besoin'''
df.keys()
```

```python
df.drop(['colourWayId', 'groupId', 'hasMultiplePrices',
       'hasVariantColours', 'id',
       'price.current.text', 'price.isMarkedDown',
       'price.isOutletPrice', 'price.previous.text', 'price.previous.value',
       'price.rrp.text', 'price.rrp.value', 'productType'], axis='columns', inplace=True)
```

```python
''' je renomme les colonnes'''
df.rename(columns={'productCode':'Product_code','name':'Name','brandName':'Brand_name','colour':'Colour', 'price.current.value':'Price_current_value','price.currency':'Price_currency','imageUrl':'Image_url','url':'Url'}, inplace=True)

df
```

```python
''' j'ordonne les colonnes'''
df =df.reindex(columns = ['Product_code','Name','Brand_name','Colour', 'Price_current_value','Price_currency','Image_url','Url'])
df
```

```python
'''Je veux enlever le nom des marques au nom du produit de la colonne Name'''
brand_list=[]
for x in df['Brand_name']:
    brand_list.append(x)
```

```python
'''Je veux récuperer chaque mot pour les supprimer de la colonne Name'''

brand_list2=[]
for x in brand_list:
    brand_list2.append(x.split())
brand_list2
```

```python
brand_list2_flat = [item for sublist in brand_list2 for item in sublist]
brand_list2_flat
```

```python
'''Je supprime les doublons de ma liste'''

brand_list2_flat = set(brand_list2_flat)
print(set(brand_list2_flat))
```

```python
'''liste des noms de produits'''

name_list=[]
for x in df['Name']:
    name_list.append(x.replace('  ',' ').replace('  ',' ').replace('  ',' '))
name_list
```

```python
'''J'ai besoin de chaque mot pour reussir a supprimer les marques'''

name_list2=[]
for x in name_list:
    name_list2.append(x.split())
name_list2
```

```python
name_list2_flat = [item.replace('-',' ') for sublist in name_list2 for item in sublist]
name_list2_flat
```

```python
'''J'applique la fonction filtre pour supprimer les noms des marques de la liste de produits'''

def word_filter(x):
    word_list = ['Moda', 'Look', 'London', 'Nursing', 'South', 'Boohoo', 'Smashbox', 'Bershka', 'Superdry', 'UGG', 'Laura', 'Jaded', 'Vintage', 'Reclaimed', 'Beach', 'Wide', 'Other', 'Lane', '&', 'Stradivarius', 'Stories', 'DESIGN', '-', 'Mama.licious', 'Missguided', 'Island', 'Urban', 'EDITION', 'Coconut', 'Hollister', '4505', 'Fit', 'Foamie', 'ASOS', 'WHITE', 'Mercier', 'Nike', 'Hope', 'Bliss', 'Pull&Bear', 'Nikki', 'PrettyLittleThing', 'Strange', 'Shiseido', 'Ivy', 'G-Star', 'River', 'Petite', 'New', 'Curve', 'Tall', 'Vero', 'Glamorous', 'Maternity', 'Running']

    if x in word_list:
        return False
    else:
        return True
    
product_name_filter =list(filter(word_filter,name_list2_flat))
product_name_filter
```

```python
''' J'applique la fonction concat space pour reformer le texte'''

from functools import reduce

def concat_space(a, b):

    c = [a,b]
    return ' '.join(c)
```

```python
product_name_string = reduce(concat_space, product_name_filter)
product_name_string
```

```python
'''Je forme à nouveau la liste de produits et je l'a prépare pour l'utiliser dans une fonction filtre'''

new_list_products = product_name_string.replace('   ','\n').replace('  ','')
print(new_list_products)
```
